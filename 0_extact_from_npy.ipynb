```python
data = np.load('images.npy', allow_pickle=True)
data_label = np.load('labels.npy', allow_pickle=True)
```

```python
import imageio
imageio.imwrite('image_name.png', data[4])
```

```python
idx = 0
for img, kilas in zip(data, data_label):
    if (kilas == 1):
        imageio.imwrite('meningioma/' + str(idx) + '.png', img)
        idx+=1
    elif (kilas == 2):
        imageio.imwrite('glioma/' + str(idx) + '.png', img)
        idx+=1
    elif (kilas == 3):
        imageio.imwrite('pituitary_tumor/' + str(idx) + '.png', img)
        idx+=1
```
